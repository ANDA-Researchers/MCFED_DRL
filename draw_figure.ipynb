```python
import matplotlib.pyplot as plt


def plot_avg_delay(
    raw_data,
    x_labels,
    line_labels,
    x_title="Number of Vehicles (%)",
    y_title="Average Delay (ms)",
    title="Average Delay Across Different Number of Vehicles",
    legend=True,
):

    data = [list(map(float, line.split())) for line in raw_data.strip().split("\n")]
    markers = [
        "o",
        "s",
        "D",
        "v",
        "^",
        "<",
        ">",
        "p",
        "P",
        "*",
        "h",
        "H",
        "+",
        "x",
        "X",
        "D",
        "d",
        "|",
        "_",
    ]

    # Plotting
    plt.figure(figsize=(8, 6))
    for i, line_data in enumerate(data):
        plt.plot(x_labels, line_data, marker=markers[i], label=line_labels[i])

    # Adding labels, legend, and title
    plt.xlabel(x_title, fontsize=14)
    plt.ylabel(y_title, fontsize=14)
    plt.title(title, fontsize=16)
    plt.xticks(x_labels, fontsize=12)
    plt.yticks(fontsize=12)
    if legend:
        plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.grid(True)

    # Show plot
    plt.show()
```

```python
# i want to draw this function with the following data
# f(x) = (lambda^x - 1) / 100 with x from 0 to 300 and lambda in [1.011, 1.012, 1.013, 1.014]

import numpy as np

x = np.linspace(0, 300, 100)
lambdas = [1.011, 1.012, 1.013, 1.014]
colors = ["r", "g", "b", "y"]

plt.figure(figsize=(8, 6))
for i, l in enumerate(lambdas):
    y = (l ** x - 1)
    plt.plot(x, y, color=colors[i], label=f"\u03BB={l}")
plt.xlabel("Total transmission power of one RSU (dBm)", fontsize=14)
plt.ylabel("Interruption Probability (%)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.grid()

plt.savefig("interruption_probability.pdf", bbox_inches="tight")
```
